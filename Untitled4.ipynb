```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score


data = pd.read_csv('creditcard.csv')

x = data.drop('Class', axis=1)  
y = data['Class']

k = 5 

accuracy_scoresrf = []
precision_scoresrf = []
recall_scoresrf = []
f1_scoresrf = []
roc_auc_scorerf = []

skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42) 

for train_index, test_index in skf.split(x, y):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

rf = RandomForestClassifier(n_estimators=100, max_depth=5)
rf.fit(x_train, y_train)

y_pred1 = rf.predict(x_test)

accuracy_scoresrf.append(accuracy_score(y_test, y_pred1))
precision_scoresrf.append(precision_score(y_test, y_pred1))
recall_scoresrf.append(recall_score(y_test, y_pred1))
f1_scoresrf.append(f1_score(y_test, y_pred1))
roc_auc_scorerf.append(roc_auc_score(y_test, y_pred1))

print("Average Accuracy:", sum(accuracy_scoresrf) / len(accuracy_scoresrf))
print("Average Precision:", sum(precision_scoresrf) / len(precision_scoresrf))
print("Average Recall:", sum(recall_scoresrf) / len(recall_scoresrf))
print("Average F1-Score:", sum(f1_scoresrf) / len(f1_scoresrf))
print("Average roc_score:", sum(roc_auc_scorerf) / len(roc_auc_scorerf))

final_model1 = RandomForestClassifier(n_estimators=100, max_depth=5) 
final_model1.fit(x, y)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc ,recall_score

fpr, tpr, thresholds = roc_curve(y_test, y_pred1) 
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'ROC Curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='No Skill')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid(True)  # Add grid for better readability
plt.show()


specificity = 1 - fpr
recalls_at_specificities = []

for thresh in thresholds:
    y_pred_thresh = (y_pred1 >= thresh).astype(int)
    recall = recall_score(y_test, y_pred_thresh)
    recalls_at_specificities.append(recall)

plt.plot(specificity, recalls_at_specificities, label='Threshold Specificity Curve')
plt.xlabel('Specificity')
plt.ylabel('Recall')
plt.title('Threshold Specificity Curve')
plt.grid(True)
plt.legend(loc="lower right")
plt.show()

```
